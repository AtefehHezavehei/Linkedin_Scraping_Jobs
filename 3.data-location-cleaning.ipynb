```python
#Comunity function
def comunidad(x):
    if x in Andalucía :
        return "Andalucía"
    elif x in Aragón :
        return "Aragón"
    elif x in Asturias :
        return "Asturias"
    elif x in Baleares:
        return "Baleares"
    elif x in Canarias :
        return "Canarias"
    elif x in Cantabria:
        return "Cantabria"
    elif x in Castilla_Mancha:
        return "Castilla_Mancha"
    elif x in Castilla_León:
        return "Castilla_León"
    elif x in Cataluña:
        return "Cataluña"
    elif x in Extremadura :
        return "Extremadura"
    elif x in Galicia:
        return "Galicia"
    elif x in Rioja:
        return "Rioja"
    elif x in Madrid :
        return "Madrid"
    elif x in Murcia :
        return "Murcia"
    elif x in Navarra :
        return "Navarra"
    elif x in País_Vasco:
        return "País_Vasco"
    elif x in Valencia :
        return "Valencia"
    elif x in Cueta_Melilla :
        return "Cueta_Melilla"
    elif x in Remote :
        return "Remote"
    else:
        return "Other"
```

```python
#Comunity Mapping
df["location"] =df["location"].map(comunidad)
```
